```

from pyspark.sql.functions import udf
```

```
def variety_pred(petal_lenght):
  if (petal_lenght <= 2):
    return 'Setosa'
  elif (petal_lenght > 3 and petal_lenght <= 5):
    return 'Versicolor'
  else:
    return 'Virginica'



iris_udf = udf(variety_pred)
iris.withColumn("variety_pred", iris_udf(iris))
```
