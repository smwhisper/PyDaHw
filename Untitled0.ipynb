```
# Для того чтобы сформулировать правило определения типа цветка я использую параметр petal_length
# судя по всему при маленьких значениях это однозначно setosa, при средних versicolor, при больших virginica
```

```

from pyspark.sql.functions import udf
```

```
def variety_pred(petal_lenght):
  if (petal_lenght <= 2):
    return 'Setosa'
  elif (petal_lenght > 3 and petal_lenght <= 5):
    return 'Versicolor'
  else:
    return 'Virginica'



iris_udf = udf(variety_pred)
iris.withColumn("variety_pred", iris_udf(iris))
```
